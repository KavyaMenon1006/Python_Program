```python
from ipyleaflet import Map,Marker,Circle,Popup,LayersControl,basemaps
```

Center the map (eg:Mumbai)

```python
center=(19.0760,72.8777)
```

Create a map

```python
Map=Map(center=center,zoom=12,basemaps=basemaps.OpenStreetMap.Mapnik)
```

Add a Marker

```python
marker=Marker(location=center,draggable=True)
Map.add_layer(marker)
```

Add layer control (lets you change basemap)

```python
Map.add_control(LayersControl(position='topright'))
```
